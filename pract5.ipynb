```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
from scipy.stats import norm
data_normal=norm.rvs(size=10000,loc=0,scale=1)
ax=sns.distplot(data_normal,bins=100,kde=True,color='skyblue',hist_kws={"linewidth":15,'alpha':1})
ax.set(xlabel="Normal Distribution",ylabel='Frequency')
```

```python
from scipy.stats import expon
data_expon=expon.rvs(size=1,loc=0,scale=1000)
ax=sns.distplot(data_expon,bins=100,kde=True,color='skyblue',hist_kws={"linewidth":15,'alpha':1})
ax.set(xlabel="Exponential Distribution",ylabel='Frequency')
```
